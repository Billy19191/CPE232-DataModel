```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_points = np.array([[5, 7], [15, 12], [16, 18], [6, 6], [16, 11], [15, 11], [6, 4], [13, 13], [16, 17], [5, 5], [15, 17], [16, 17], [6, 8], [14, 12], [16, 15]])
centroids = np.array([[5, 12], [15, 4], [10, 17]])
```

```python
def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

def assign_to_clusters(data_points, centroids):
    clusters = {}
    distances = []
    for point in data_points:
        distances_to_centroids = [euclidean_distance(point, centroid) for centroid in centroids]
        distances.append(distances_to_centroids)
        cluster_index = np.argmin(distances_to_centroids)
        if cluster_index not in clusters:
            clusters[cluster_index] = [point]
        else:
            clusters[cluster_index].append(point)
    return clusters, distances

def update_centroids(clusters):
    new_centroids = []
    for key in clusters:
        new_centroids.append(np.mean(clusters[key], axis=0))
    return np.array(new_centroids)
```

```python
for iteration in range(3):
    print("Iteration", iteration+1)
    clusters, distances = assign_to_clusters(data_points, centroids)
    centroids = update_centroids(clusters)
    print("Updated Centroids:", centroids)
    print("Distances:")
    for i, point in enumerate(data_points):
        print("Point:", point, "| Distances:", distances[i], "| Cluster:", np.argmin(distances[i])+1)
    colors = ['r', 'g', 'b']  # Colors for each cluster
    plt.figure(figsize=(8, 6))
    for key in clusters:
        cluster_points = np.array(clusters[key])
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], color=colors[key], label=f'Cluster {key+1}')
    plt.scatter(centroids[:, 0], centroids[:, 1], color='orange', marker='*', s=200, label='Centroids')
    plt.title(f'K-Means Clustering - Iteration {iteration+1}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid(True)
    plt.show()
print("Final clusters:")
for key in clusters:
    print("Cluster", key+1, ":", clusters[key])
```
